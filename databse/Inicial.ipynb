```python
import sqlite3
```

```python
# Conectar ao banco de dados
conn = sqlite3.connect("document_comparisons.db")
cursor = conn.cursor()

# Buscar todos os dados da tabela
cursor.execute("SELECT * FROM Comparacoes;")
rows = cursor.fetchall()

print("Dados na tabela 'Comparacoes':")
for row in rows:
    print(row)

conn.close()
```

```python
# Conectar ao banco de dados
conn = sqlite3.connect("document_comparisons.db")
cursor = conn.cursor()

# Verificar a estrutura da tabela "Comparacoes"
cursor.execute("PRAGMA table_info(Comparacoes);")
columns = cursor.fetchall()

print("Estrutura da tabela 'Comparacoes':")
for column in columns:
    print(column)

conn.close()
```
