```python
import requests

# URL base da API
BASE_URL = "https://metadata.idl.ucsf.edu/solr/ltdl3/query"

def fetch_types_for_collection(collection_name, max_results=100):
    """
    Busca todos os tipos de documentos para uma coleção específica.
    
    Args:
        collection_name (str): Nome da coleção.
        max_results (int): Número máximo de documentos por página.
    
    Returns:
        set: Conjunto de tipos de documentos encontrados.
    """
    unique_types = set()
    start = 0

    while True:
        params = {
            "q": f"collection:\"{collection_name}\"",
            "fl": "type",
            "wt": "json",
            "rows": max_results,
            "start": start
        }

        response = requests.get(BASE_URL, params=params)
        if response.status_code == 200:
            docs = response.json().get("response", {}).get("docs", [])
            if not docs:
                break  # Sai do loop se não houver mais documentos

            # Adiciona os tipos únicos ao conjunto
            for doc in docs:
                doc_types = doc.get("type")
                if isinstance(doc_types, list):
                    unique_types.update(doc_types)
                elif doc_types:
                    unique_types.add(doc_types)

            start += max_results
        else:
            print(f"Erro ao buscar tipos: {response.status_code} - {response.text}")
            break

    return unique_types

# Busca os tipos de documentos para a coleção "Actos Litigation Documents"
collection_name = "Actos Litigation Documents"
types = fetch_types_for_collection(collection_name)

print(f"Tipos de documentos para a coleção '{collection_name}':")
print(sorted(types))
```

```python
import requests

# URL base da API
BASE_URL = "https://metadata.idl.ucsf.edu/solr/ltdl3/query"

def fetch_document_count_for_collection(collection_name):
    """
    Busca a quantidade total de documentos para uma coleção específica.

    Args:
        collection_name (str): Nome da coleção.

    Returns:
        int: Número total de documentos na coleção.
    """
    query = f"collection:\"{collection_name}\""
    params = {
        "q": query,
        "wt": "json",
        "rows": 0  # Apenas queremos o número total de documentos
    }

    response = requests.get(BASE_URL, params=params)
    if response.status_code == 200:
        return response.json().get("response", {}).get("numFound", 0)
    else:
        print(f"Erro na API: {response.status_code} - {response.text}")
        return 0

# Exemplo de uso
collection_name = "Actos Litigation Documents"
document_count = fetch_document_count_for_collection(collection_name)
print(f"Quantidade de documentos na coleção '{collection_name}': {document_count}")
```
